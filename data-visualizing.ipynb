```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
file_path = "../input/fivethirtyeight-comic-characters-dataset/dc-wikia-data.csv"
comic_data = pd.read_csv(file_path,index_col='page_id',parse_dates=True)
```

```python
comic_data.head()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=comic_data)
```

```python
list(comic_data.columns)
```

```python
plt.figure(figsize=(16,6))
plt.title("comic movie")
sns.barplot(x=comic_data['ID'],y=comic_data["APPEARANCES"])
plt.ylabel("appearance")
```

```python
sns.scatterplot(x=comic_data['YEAR'],y=comic_data['APPEARANCES'])
```

```python
sns.scatterplot(x=comic_data['YEAR'],y=comic_data['APPEARANCES'],hue=comic_data['SEX'])
```

```python
sns.kdeplot(data=comic_data['APPEARANCES'],shade=True)
```
