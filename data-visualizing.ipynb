```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
file_path = "../input/fivethirtyeight-comic-characters-dataset/dc-wikia-data.csv"
comic_data = pd.read_csv(file_path)
```

```python
comic_data.head()
```

```python
comic_data['DC Comic']='DC'
```

```python
comic_data.head()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=comic_data)
```

```python
plt.figure(figsize=(14,6))
plt.title("comic movie")
sns.lineplot(data=comic_data['APPEARANCES'],label="DC Comic")
plt.xlabel("appearance")
```

```python
plt.figure(figsize=(16,6))
plt.title("comic movie")
sns.barplot(x=comic_data['ID'],y=comic_data["APPEARANCES"])
plt.ylabel("appearance")
```

```python
sns.scatterplot(x=comic_data['YEAR'],y=comic_data['APPEARANCES'])
```

```python
sns.scatterplot(x=comic_data['YEAR'],y=comic_data['APPEARANCES'],hue=comic_data['SEX'])
```

```python
sns.kdeplot(data=comic_data['APPEARANCES'],shade=True)
```

```python
sns.jointplot(x=comic_data['APPEARANCES'],y=comic_data['YEAR'],kind='kde')
```

```python
file_m_path='../input/fivethirtyeight-comic-characters-dataset/marvel-wikia-data.csv'
comic_m_data= pd.read_csv(file_m_path)
```

```python
comic_m_data.head()
```

```python
comic_m_data['MC Comic']='MC'
```

```python
comic_m_data.head()
```

```python
plt.figure(figsize=(16,6))
sns.distplot(a=comic_data['APPEARANCES'],label='DC Comic',kde=False)
sns.distplot(a=comic_m_data['APPEARANCES'],label='MC Comic',kde=False)
plt.title("Comic Movies of DC and MC")
plt.legend()
```

```python
plt.figure(figsize=(16,6))
sns.kdeplot(data=comic_data['APPEARANCES'],label='lable',shade=True)
sns.kdeplot(data=comic_m_data['APPEARANCES'],label='lable',shade=True)
plt.title("Comic Movies of DC and MC")
```
